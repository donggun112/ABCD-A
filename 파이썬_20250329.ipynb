```python
p_class = "Z"
sel_amount = 79900

if p_class == "A":
  sel_amount *= 0.7
  print(f'판매가는 {sel_amount}원 입니다.')
elif p_class == "B":
  sel_amount *= 0.85
  print(f'판매가는 {sel_amount}원 입니다.')
elif p_class == "C":
  sel_amount *= 0.92
  print(f'판매가는 {sel_amount}원 입니다.')
elif p_class == "Z":
  sel_amount += 5000
  print(f'판매가는 {sel_amount}원 입니다.')
```

```python
i = 0
result1 = 0
while i < 100:
    i = i + 1
    if i % 2 == 0:
        print('1번 방법 : {0} {1}'.format(i,result1))
        result1 = result1 + i

print('1번 방법 : {0}'.format(result1))
```

```python
j = 0
result2 = 0
while True:
    if j > 100:
        break

    j = j + 1
    if j % 2 == 0:
        result2 = result2 + j
        print('2번 방법 (break) : {0}'.format(result2))
```
